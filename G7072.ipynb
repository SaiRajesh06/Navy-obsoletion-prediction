```python
#STATS

import numpy as np
from scipy.stats import mode

def temporalize(data, lookback):
    output_data = []
    for i in range(data.shape[1]):
        column = data[:, i]
        column = column.astype(float)
        temp_column = []
        for j in range(lookback, len(column)):
            window = column[j-lookback:j]
            mean = np.mean(window)
            median = np.median(window)
            min_val = np.min(window)
            max_val = np.max(window)
            mode_result = mode(window)
            if isinstance(mode_result[0], np.ndarray):
                mode_val = mode_result[0][0]
            else:
                mode_val = mode_result[0]
            std = np.std(window)
            mad = np.mean(np.abs(window - np.mean(window)))
            mse = np.mean((window - np.mean(window))**2)
            abs_dev = np.mean(np.abs(window - np.mean(window)))
            mad_mean = np.mean(np.abs(window - np.median(window)))
            temp_column.append([mean, median, min_val, max_val, mode_val, std, mad, mse, abs_dev, mad_mean])
        output_data.append(temp_column)
    return output_data

lookback = 10
temporalized_data = temporalize(differences.values, lookback)

for i, column in enumerate(temporalized_data):
    print(f"Column {i+1}:")
    for j, stats in enumerate(column):
        print(f"Step {j}: {stats}")
    print()
```

```python
import pandas as pd

def label_columns(df, threshold=1.5):
    labeled_df = df.copy()
    for col in df.columns:
        col_values = df[col].astype(float)
        if (col_values < threshold).any():
            labeled_df[col] = col_values.apply(lambda x: 1)
        else:
            labeled_df[col] = col_values.apply(lambda x: 0)
    return labeled_df

# Assuming 'selected_values' is your original DataFrame
labeled_values = label_columns(selected_values.iloc[1:])
print(labeled_values)
```

```python
import numpy as np
import pandas as pd

def temporalize(data, lookback):
    output_data = []
    # Iterate over each column in the data
    for i in range(data.shape[1]):
        column = data[:, i]
        temp_column = []
        # Create temporalized sequences for the current column
        for j in range(lookback, len(column)):
            temp_column.append(column[j-lookback:j])
        output_data.append(temp_column)
    return output_data

# Assuming labeled_values is a pandas DataFrame
labeled_values_array = labeled_values.values
lookback = 10
# Temporalize the data
y_label = temporalize(labeled_values_array, lookback)

# Print the temporalized data
for i, column in enumerate(y_label):
    print(f"Column {i+1}:")
    for j, step in enumerate(column):
        print(f" {step}")
    print()
```

```python
# Constructing X with only statistical values
X = []
for column in temporalized_data:
    for stats in column:
        X.append(stats)

# Printing X in the specified format
print("X = [")
for stats in X:
    print(stats)
print("]")

```

```python

X = np.array(X)
y = np.array(y)

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X = np.array(X)
y = np.array(y)

trainX, testX, trainY, testY = train_test_split(X, y, test_size=1/3, random_state=42)

print("X_train shape:", trainX.shape)
print("X_test shape:", testX.shape)
print("y_train shape:", trainY.shape)
print("y_test shape:", testY.shape)
```

```python
print("TrainX shape:", trainX)
print("testX shape:", testX)
print("Trainy shape:", trainY)
print("Testy shape:", testY)
```

```python

import numpy as np
from keras.models import Model
from keras.layers import Input, LSTM, Dense, RepeatVector, TimeDistributed, Dropout
from keras.models import Sequential

timesteps = 10
n_features = 1

input_shape = (timesteps, n_features)

autoencoder = Sequential([
    Input(shape=input_shape),
    LSTM(64, return_sequences=True),
    LSTM(32, return_sequences=True),
    LSTM(64, return_sequences=True),
    Dense(n_features, activation='sigmoid')
])

autoencoder.compile(optimizer='adam', loss='mean_squared_error')

autoencoder.fit(trainX, trainX, epochs=10, batch_size=32, validation_split=1/3)

encoder = Model(inputs=autoencoder.input, outputs=autoencoder.layers[2].output)

classification_model = Sequential([
    encoder,
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

classification_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

classification_model.fit(trainX, trainY, epochs=10, batch_size=32, validation_split=1/3)

predictions = classification_model.predict(testX)

binary_predictions = np.round(predictions).astype(int)

print("Predicted binary values:", binary_predictions)

```
